```python
class BankAccount:
    def __init__(self, owner, password, balance=0):
        self.owner = owner
        self.password = password
        self.balance = balance
        self.deposit_limit = 1000
        self.withdraw_limit = 500

    def deposit(self, amount):
        if amount <= 0:
            print("Geçerli bir miktar giriniz.")
        elif amount > self.deposit_limit:
            print(f"En fazla {self.deposit_limit} TL yatırabilirsiniz.")
        else:
            self.balance += amount
            print(f"{amount} TL yatırıldı. Yeni bakiye: {self.balance} TL")

    def withdraw(self, amount):
        if amount <= 0:
            print("Geçerli bir miktar giriniz.")
        elif amount > self.withdraw_limit:
            print(f"En fazla {self.withdraw_limit} TL çekebilirsiniz.")
        elif amount > self.balance:
            print("Yetersiz bakiye.")
        else:
            self.balance -= amount
            print(f"{amount} TL çekildi. Kalan bakiye: {self.balance} TL")

    def show_balance(self):
        print(f"{self.owner} adlı hesabın mevcut bakiyesi: {self.balance} TL")

# --- Ana Program ---

# Hesap oluşturuluyor (şifre sistemde kayıtlı)
account = BankAccount("Kullanıcı", "a1a2a3a4", 500)

# Kullanıcıdan şifre alınır
input_password = input("Lütfen şifrenizi giriniz: ")

if input_password == account.password:
    print("\nGiriş başarılı!\n")
    
    while True:
        print("\n--- İşlem Menüsü ---")
        print("1. Bakiye Görüntüle")
        print("2. Para Yatır")
        print("3. Para Çek")
        print("4. Çıkış")
        
        choice = input("Bir işlem seçiniz (1-4): ")

        if choice == "1":
            account.show_balance()
        elif choice == "2":
            amount = float(input("Yatırmak istediğiniz miktarı girin: "))
            account.deposit(amount)
        elif choice == "3":
            amount = float(input("Çekmek istediğiniz miktarı girin: "))
            account.withdraw(amount)
        elif choice == "4":
            print("Çıkış yapılıyor. İyi günler!")
            break
        else:
            print("Geçersiz seçim. Lütfen 1-4 arasında bir sayı girin.")
else:
    print("\nHatalı şifre! İşlemlere devam edemezsiniz.")
```
