# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız

import time
def vki_hesaplama(kilo,boy):
    vki = kilo/(boy**2)
    return round(vki,1)
def deger_araligi(vki):
    if vki < 18.5:
        return "zayıf"
    elif vki <=24.9:
        return "normal kilolu"
    elif vki <=29.9:
        return "fazla kilolu"
    elif vki <= 39.9:
        return "obez"
    elif vki>=40:
        return "ileri obez"
    
def cikti(isim, yas, kilo, boy, vki, deger):
    kayit_tarihi = time.ctime(time.time())
    t1 = f"kullanici adi: {isim}\n"
    t2 = f"kullanicinin yas, kilo, boy ilgisi: {yas}, {kilo}, {boy}\n"
    t3 = f"kullanicini vucut kitle indexi: {vki}\n"
    t4 = f"deger araligi: {deger}\n"
    t5 = f"olusturma tarihi: {kayit_tarihi}\n"
    output = t1 + t2 + t3 + t4 + t5
    return output

def main():
    giris_sayisi = int(input("kac kullanici verisi girilecektir?"))
    for i in range(giris_sayisi):
        print(f"kullanici kaydi {i+1}")
        isim = input("isim soyisim: ")
        yas = int(input("yas :"))
        kilo = float(input("kilo: "))
        boy = float(input("boy (m): "))
        vki = vki_hesaplama(kilo, boy)
        deger = deger_araligi(vki)
        output = cikti(isim, yas, kilo, boy, vki, deger)
        with open("saglik.txt","a") as f:
            f.write(output)
            f.write("-"*20)
            f.write("\n")
            print("kullanici kaydedildi.")
    print("veriler 'saglik.txt' dosyasina kaydedildi.\n")

main()



    
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız

import time
menu = {
    "elma" : 0.50,
    "armut" : 0.80,
    "cilek" : 0.65,
    "muz" : 1.75,
    "kavun" : 2.15,
    "karpuz" : 2.05    
}

def display_menu():
    print("----yetbar menu-----")
    for i in list(menu.keys()):
        print(f"{i.upper():<10} : {menu[i]} TL")
def tekUrunFiyatHesaplama(urun, adet):
    assert adet >= 0, "urun adedi 0 veya 0'dan kücük olamaz"
    if urun not in list(menu.keys()):
        return None
    else:
        return menu[urun] * adet

def satin_alma():
    display_menu()
    print("lütfen satin almak istediginiz urunu giriniz. cikis yapmak icin 0'i tuslayiniz.")
    satinAlinanUrunler = []
    satinAlmaAdedi = []
    sonFiyat = 0
    while 1:
        eklenecekUrun = input("hangi urunden eklemek istersiniz?")
        if eklenecekUrun == "0":
            print("fis kaydediliyor")
            break
        eklemeAdedi = int(input("kac adet eklemek istiyorsunuz?"))
        fiyat = tekUrunFiyatHesaplama(eklenecekUrun, eklemeAdedi)
        if fiyat is None:
            print("yanlis bir giris yaptiniz, urun eklenemedi") 
        else: 
            print(f"satin alma ucreti: {fiyat}.\n Satin almayi onaylamak icin 'y' iptal etmek icin 'n' giriniz.")
            check = input("satin almayi onayliyor musunuz?")
            if check == "y":
                print("satin alma onaylandi.")
                print("-"*40)
                sonFiyat += fiyat
                satinAlinanUrunler.append(eklenecekUrun)
                satinAlmaAdedi.append(eklemeAdedi)
                print(f"ara toplam: {sonFiyat}")
            elif check == "n":
                print("satin alma islemi iptal edildi")
                print("-"*40)
    return sonFiyat, satinAlinanUrunler, satinAlmaAdedi

def main():
    print("yetbar a hosgeldiniz. menuyu goruntulemek icin 1'e, smoothie yapmak icin 2'ye basiniz. cikmak icin 0'a basiniz.")
    program = None
    while 1:
        program = int(input("tercihiniz: "))
        if program < 0 or program > 2:
            print("yanlis bir tercih yaptiniz. tekrar deneyiniz.")
        elif program == 0:
            print("cikis yapiliyor")
            break
        elif program == 1:
            display_menu()
        elif program == 2:
            satinAlmaUcreti, satinAlinanUrunler, satinAlmaAdedi = satin_alma()
            with open("kasa.txt","a") as f:
                f.write(str(time.ctime(time.time())))
                f.write("\n")
                f.write(",".join(map(str, satinAlinanUrunler)))
                f.write("\n")
                f.write(",".join(map(str, satinAlmaAdedi)))
                f.write("\n")
                f.write(f"toplam: {satinAlmaUcreti}")
                f.write("\n")
                f.write("-"*20)
                f.write("\n")
            cikisKontrol = input("baska bir islem yapmak istiyor musunuz ( evet-hayir)")
            if cikisKontrol == "evet":
                pass
            else:
                print("\n\n TESEKKURLER")
                break

main()



```
