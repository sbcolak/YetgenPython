```python
import datetime

# Menü sözlüğü, kahve çeşitleri, süt çeşitleri ve şeker fiyatını içerir.
menu = {
    "kahveler": {
        "latte": 12.50,
        "cappuccino": 19.80,
        "americano": 9.65,
        "cold brew": 10.75,
        "flat white": 19.15,
        "filtre kahve": 8.05,
    },
    "sutler": {
        "normal süt": 0,
        "badem sütü": 1.75,
        "laktozsuz süt": 1.75,
        "yulaf sütü": 2.05,
    },
    "seker": 0.25
}

# Kahve makinesi sınıfı oluşturuluyor
class CoffeeMachine:
    def __init__(self, menu):
        self.menu = menu  # Menü sözlüğü sınıfa tanımlanıyor

    # Kahve seçeneklerini gösterir
    def kahve_menu_goster(self):
        print("\n--- KAHVE ÇEŞİTLERİ ---")
        for idx, (kahve, fiyat) in enumerate(self.menu['kahveler'].items(), 1):
            print(f"{idx}. {kahve.capitalize():<15} : {fiyat} TL")

    # Süt seçeneklerini gösterir
    def sut_menu_goster(self):
        print("\n--- SÜT SEÇENEKLERİ ---")
        for idx, (sut, fiyat) in enumerate(self.menu['sutler'].items(), 1):
            extra = f"(+{fiyat} TL)" if fiyat > 0 else "(ücretsiz)"
            print(f"{idx}. {sut.capitalize():<15} : {extra}")

    # Kahve seçimini kullanıcıdan alır
    def kahve_sec(self):
        kahve_secim = int(input("\nKahve numarasını seçiniz: "))
        kahve_listesi = list(self.menu['kahveler'].keys())
        return kahve_listesi[kahve_secim - 1]

    # Süt seçimini kullanıcıdan alır
    def sut_sec(self):
        sut_secim = int(input("Süt numarasını seçiniz: "))
        sut_listesi = list(self.menu['sutler'].keys())
        return sut_listesi[sut_secim - 1]

    # Şeker miktarını kullanıcıdan alır
    def seker_sec(self):
        return int(input("Kaç adet şeker eklemek istersiniz (0-5): "))

    # Ödeme alır ve kontrol eder
    def odeme_yap(self, toplam_tutar):
        verilen_para = float(input("Lütfen para ekleyiniz: "))
        while verilen_para < toplam_tutar:
            eksik = toplam_tutar - verilen_para
            print(f"Yetersiz bakiye, lütfen {eksik:.2f} TL daha ekleyin.")
            verilen_para += float(input("Ek para giriniz: "))
        return verilen_para - toplam_tutar

    # Kullanıcıdan sipariş alır ve toplam tutarı hesaplar
    def siparis_al(self):
        self.kahve_menu_goster()
        kahve = self.kahve_sec()

        self.sut_menu_goster()
        sut = self.sut_sec()

        seker_adet = self.seker_sec()

        toplam_tutar = (self.menu['kahveler'][kahve] +
                        self.menu['sutler'][sut] +
                        seker_adet * self.menu['seker'])

        print(f"\nToplam tutar: {toplam_tutar:.2f} TL")

        para_ustu = self.odeme_yap(toplam_tutar)

        print(f"\nSiparişiniz alındı. Para üstünüz: {para_ustu:.2f} TL")
        print("Kahveniz hazırlanıyor... Afiyet olsun!")

        self.log_kaydet(kahve, sut, seker_adet, toplam_tutar)

    # Siparişleri tarih ve saat ile log dosyasına kaydeder
    def log_kaydet(self, kahve, sut, seker, tutar):
        zaman = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        log_mesaj = f"{zaman} - {kahve.capitalize()}, {sut}, {seker} şeker, Toplam: {tutar:.2f} TL\n"
        with open("siparis_log.txt", "a", encoding="utf-8") as log:
            log.write(log_mesaj)

# Kullanım örneği
kahve_otomat = CoffeeMachine(menu)
kahve_otomat.siparis_al()
```
